```python
from PIL import Image

# 이미지 불러오기
img = Image.open('티모대위.jpg')
```

```python
# 객체 (object)
# 변수와 명령어(함수)를 같이 가지는 데이터

# 이미지의 크기
img.size

# 튜플 : 수정이 안되는 리스트
# img.size[0]
img.size[1]
```

```python
img.width # 이미지 너비
```

```python
img.height # 이미지 높이
```

```python
# 이미지 크기 변경
img_resized = img.resize((320, 200))
img_resized.save('티모대위(수정).jpg')
```

```python
# 이미지 자르기
img.crop((0, 0, 300, 200))
```

```python
# 이미지 회전
img.rotate(90)
```

```python
# 이미지 대칭
img.transpose(Image.FLIP_LEFT_RIGHT) # 좌우 대칭
img.transpose(Image.FLIP_TOP_BOTTOM) # 상하 대칭
```

```python
# 이미지 필터
from PIL import ImageFilter

# img.filter(ImageFilter.BLUR)
# img.filter(ImageFilter.BoxBlur(10))
img.filter(ImageFilter.GaussianBlur(10))
```

```python
# 특정 부분만 블러 처리 하기

# 1. 이미지 자르기
cropped_img = img.crop((300, 200, 600, 400))

# 2. 자른 이미지를 블러처리
blured_img = cropped_img.filter(ImageFilter.GaussianBlur(10))

# 3. 원본에 블러처리한 이미지를 붙여넣기
img.paste(blured_img, (300, 200, 600, 400))
img.save('티모대위(부분수정).jpg')
```
