```python
import xlwings as xw
import pandas as pd

data_file = 'order_data/주문내역_샘플.xlsx'
data_file_ref = 'order_data/주문내역_참조데이터.xlsx'

# 주문내역_샘플을 읽어서 데이터 프레임으로 변경
# df 변수에 저장
wb = xw.Book(data_file)
ws = wb.sheets['Sheet1']
df = ws.range('A1').options(pd.DataFrame, expand='table', index=False).value
df
```

```python
# 주문내역_참조데이터를 읽어서 데이터 프레임으로 변경
# df_ref 변수에 저장
wb = xw.Book(data_file_ref)
ws = wb.sheets['Sheet1']
df_ref = ws.range('A1').options(pd.DataFrame, expand='table', index=False).value
df_ref
```

```python
# 두 개의 데이터프레임 병합 
df_new = df.merge(df_ref)
df_new
```

```python
df_new = df_new[['주문번호', '제품명', '제품코드', '제품가격', '수량', '발주처']]
df_new
```

```python
df_new.to_excel('order_data/주문내역_완성본.xlsx', index=False)
```

```python
# merge 개념 이해하기
df_left = pd.DataFrame({
    'key' : ['k0', 'k1', 'k2', 'k3'],
    'value1' : ['a0', 'a1', 'a2', 'a3']
})
df_left
```

```python
df_right = pd.DataFrame({
    'key' : ['k2', 'k3', 'k4', 'k5'],
    'value2' : ['b0', 'b1', 'b2', 'b3']
})
df_right
```

```python
df_left.merge(df_right, how='left', on='key')
```

```python
df_left.merge(df_right, how='right', on='key')
```

```python
df_left.merge(df_right, how='outer', on='key')
```

```python
df_left.merge(df_right, how='inner', on='key') # df_left.merge(df_right) 와 동일
```
