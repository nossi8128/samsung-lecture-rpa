```python
import os
import xlwings as xw
import pandas as pd
# 데이터 폴더 경로
folder_path = r'C:\samsung_auto2\02.엑셀자동화\02.xlwings\onlinesale_data'

file_list = os.listdir(folder_path)

# 빈 데이터 프레임 생성
total_df = pd.DataFrame()
for file in file_list:
    excel_path = os.path.join(folder_path, file)
    
    # 엑셀 불러오기
    wb = xw.Book(excel_path)

    # 시트 선택
    ws = wb.sheets['data']

    # A1부터 표 데이터를 읽어 데이터프레임으로 변경
    df = ws.range('A1').options(pd.DataFrame, expand='table', index=False).value
    
    # 데이터 프레임 이어주기
    total_df = pd.concat([total_df, df], ignore_index=True)

    # 엑셀 닫기
    xw.apps.active.quit()

# 합계 파일 저장
total_df.to_excel(f'{folder_path}/total.xlsx', index=False)
```

```python
# 자동 열 너비 맞추기
wb = xw.Book('onlinesale_data/total.xlsx')
ws = wb.sheets['Sheet1']
ws.autofit(axis='columns') # myrange.columns.autofit()
wb.save('onlinesale_data/total.xlsx')
xw.apps.active.quit()
```
