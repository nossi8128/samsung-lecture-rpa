```python
import pandas as pd
df = pd.read_excel('member.xlsx', index_col='번호')
df
```

```python
df.sort_values('체중') # ascending=True 기본값 : 오름차순
```

```python
df.sort_values('체중', ascending=False) # 내림차순
```

```python
df.sort_values(['출생연도', '체중'])
```

```python
# 새로운 row 추가
new_row = {
    '이름' : '지민',
    '출생연도' : 1995,
    '출생지' : '부산',
    '담당파트' : '메인댄서',
    '체중' : 58
}
df2 = pd.DataFrame(new_row, index=['8번'])
df = pd.concat([df, df2])
df
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
new_row = {
    '이름' : '스타트코딩'
}
df2 = pd.DataFrame(new_row, index=['8번'])
df = pd.concat([df, df2])
df
```

```python
# Nan 을 가지는 행을 버림
df.dropna()
```

```python
df.dropna(axis=1) # 열을 기준으로 삭제
```

```python
df.dropna(how='all')
```

```python
df.dropna(subset=['이름', '출생지'])
```

```python
# Nan 을 가지는 0으로 대체
df.fillna(0)
```
